```python
import fitz  # PyMuPDF
import requests

# URL of the PDF (ensure it's a direct link to the PDF)
pdf_url = 'https://github.com/rsetaung/pdf/raw/59928e93cab03f0391f2c9132febe6087692b086/Sample1_National_Directory_MH_facilities_Cleaned_AL_1.pdf'

try:
    # Send an HTTP GET request to download the PDF
    response = requests.get(pdf_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Create a PDF document object from the downloaded PDF content
        pdf_document = fitz.open(stream=response.content, filetype="pdf")

        # Initialize a variable to store extracted text
        extracted_text = ""

        # Iterate through pages and extract text
        for page_num in range(len(pdf_document)):
            page = pdf_document[page_num]
            page_text = page.get_text()
            extracted_text += page_text

        # Print or process the extracted text for NLP
        print(extracted_text[:500])  # Print the first 500 characters as a sample

    else:
        print(f'Failed to download PDF. Status code: {response.status_code}')

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re

# Split the text into paragraphs
paragraphs = re.split(r'\n\s*\n', extracted_text)

# Initialize variables to store extracted information
names = []
addresses = []
cities = []
states = []
zips = []
services = []

# Iterate through paragraphs to extract information
for i in range(len(paragraphs)):
    if i % 2 == 0:  # Odd paragraphs
        lines = paragraphs[i].split('\n')
        # Check if there are enough lines to extract information
        if len(lines) >= 4:
            # Extract name
            name = lines[0].strip()
            names.append(name)
            # Extract address, city, and state
            address = lines[1].strip()
            city = lines[2].strip()
            state = lines[3].strip()
            zip_code = re.search(r'\b\d{5}\b', address)
            if zip_code:
                zips.append(zip_code.group())
            else:
                zips.append('')
            addresses.append(address)
            cities.append(city)
            states.append(state)
    else:  # Even paragraphs
        services_text = paragraphs[i].strip()
        # Extract services
        services.append(services_text)

# Print the extracted information
for i in range(len(names)):
    print(f"Name: {names[i]}")
    print(f"Address: {addresses[i]}")
    print(f"City: {cities[i]}")
    print(f"State: {states[i]}")
    print(f"Zip: {zips[i]}")
    print(f"Services: {services[i]}")
    print("\n")
```
